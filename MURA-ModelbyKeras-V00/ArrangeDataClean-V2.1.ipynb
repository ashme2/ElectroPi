### 1. Creating ImagePathes-Labels.csv
***

##### This code arranges/divides MURA Dataset (Elbow - Finger - Forearm - Hand - Humerus - Shoulder - wrist) to create ImagesPathes-Labels.csv which will contain 3 columns without column headers or row index as the following sample:
##### (the headers/index only for explanation)
|Row Index|Label for Normality|Image Path|Bone Type|

and so on

```python
import pandas as pd
```

```python
def ArrData(dirPath, ImageCSVName, LabelCSVName):
    print('\n [INFO] laoding csv file...')
    ImageFllPath=dirPath + "\\" + ImageCSVName                         #Image Full Path
    Image=pd.read_csv(ImageFllPath, names= ["Image Path"])
    print(ImageFllPath)
    Label=[]                                                           #Label for Normality List
    Type=[]                                                            #Bone Type List
    IndexNo=[50000, 0, 50000, 0, 50000, 0, 50000, 0, 50000, 0, 50000, 0, 50000, 0]  #Index List
    #IndexNo[0]= 50000                                                 #Elbow Start Index
    #IndexNo[1]= 0                                                     #Elbow End Index
    #IndexNo[2]= 50000                                                 #Finger Start Index
    #IndexNo[3]= 0                                                     #Finger End Index
    #IndexNo[4]= 50000                                                 #Forearm Start Index
    #IndexNo[5]= 0                                                     #Forearm End Index
    #IndexNo[6]= 50000                                                 #Hand Start Index
    #IndexNo[7]= 0                                                     #Hand End Index
    #IndexNo[8]= 50000                                                 #Humerus Start Index
    #IndexNo[9]= 0                                                     #Humerus End Index
    #IndexNo[10]= 50000                                                #Shoulder Start Index
    #IndexNo[11]= 0                                                    #Shoulder End Index
    #IndexNo[12]= 50000                                                #Wrist Start Index
    #IndexNo[13]= 0                                                    #Wrist End Index
    i=0
    print('\n [INFO] creating data table...')
    # i from 0 to length of Image -1
    for i in range(len(Image)):
        x=Image.iloc[i,0]
        # The no. of letters (pos/neg) from right  direction is (19)
        # MURA-v1.1/train/XR_SHOULDER/patient00001/study1_positive/image1.png
        # Or The no. of letters (pos/neg) from right  direction is (20)
        # MURA-v1.1/train/XR_SHOULDER/patient00001/study1_positive/image10.png
        ts1=x[-6:-1]                                                 # Testing Image have Single Digit No. or Double Digit No.
        ts2=ts1[0:2]
        if ts2.isnumeric():
            st1=x[-20:-1]
        else:
            st1=x[-19:-1]
        st2=st1[0:3]
        if st2=="pos":
            Label.insert(i, 1)                                        # Adding 1 for Positive (Abnormal)
        else:
            Label.insert(i, 0)                                        # Adding 0 for Negative (Normal)
        # The no. of letters (elBOW/finGER/foreARM/hAND/humeRUS/shoulDER/wrIST) from right  direction is (43)
        # MURA-v1.1/train/XR_SHOULDER/patient00001/study1_positive/image1.png
        # The no. of letters (elBOW/finGER/foreARM/hAND/humeRUS/shoulDER/wrIST) from right  direction is (44)
        # MURA-v1.1/train/XR_SHOULDER/patient00001/study1_positive/image10.png
        if ts2.isnumeric():
            st3=x[-44:-1]
        else:
            st3=x[-43:-1]
        st4=st3[0:3]
        if st4=="BOW":
            Type.insert(i, "Elbow")
            if i<IndexNo[0]:
                IndexNo[0]=i
            if i>IndexNo[1]:
                IndexNo[1]=i
        elif st4=="GER":
            Type.insert(i, "Finger")
            if i<IndexNo[2]:
                IndexNo[2]=i
            if i>IndexNo[3]:
                IndexNo[3]=i
        elif st4=="ARM":
            Type.insert(i, "Forearm")
            if i<IndexNo[4]:
                IndexNo[4]=i
            if i>IndexNo[5]:
                IndexNo[5]=i
        elif st4=="AND":
            Type.insert(i, "Hand")
            if i<IndexNo[6]:
                IndexNo[6]=i
            if i>IndexNo[7]:
                IndexNo[7]=i
        elif st4=="RUS":
            Type.insert(i, "Humerus")
            if i<IndexNo[8]:
                IndexNo[8]=i
            if i>IndexNo[9]:
                IndexNo[9]=i
        elif st4=="DER":
            Type.insert(i, "Shoulder")
            if i<IndexNo[10]:
                IndexNo[10]=i
            if i>IndexNo[11]:
                IndexNo[11]=i
        else:
            Type.insert(i, "Wrist")
            if i<IndexNo[12]:
                IndexNo[12]=i
            if i>IndexNo[13]:
                IndexNo[13]=i
    
    print('\n [INFO] saving data file...')
    LabelFllPath=dirPath + "\\ImagePathes-" + LabelCSVName             #Label Full Path
    df= pd.DataFrame(Label)                                            #Convert Label List to DataFrame
    dfType= pd.DataFrame(Type)                                         #Convert Type List to DataFrame
    df.insert(loc=1, column='Path', value=Image)                       #Insert Image DataFrame to Label
    df.insert(loc=2, column='Type', value=dfType)                      #Insert Type DataFrame to Label
    
    # header: for column name, index: for row name
    df.to_csv(LabelFllPath, sep=',', header= False, index=False)       #Save Label and Image as .csv file
    
    IndexFllPath=dirPath + "\\Index-" + LabelCSVName                   #Index Full Path
    dfIndex= pd.DataFrame(IndexNo)                                     #Convert Index List to DataFrame
    # header: for column name, index: for row name
    dfIndex.to_csv(IndexFllPath, sep=',', header= False, index=False)  #Save Index as .csv file
    
```

```python
#Arrange Train Data
Path='C:\\Users\\Moris\\MURA Code'
ImNm='train_image_paths.csv'
LbNm='train_labeled_studies.csv'
ArrData(Path, ImNm, LbNm)

#Arrange Test/Valid Data
Path='C:\\Users\\Moris\\MURA Code'
ImNm='valid_image_paths.csv'
LbNm='valid_labeled_studies.csv'
ArrData(Path, ImNm, LbNm)
```

|Trn: Train Image|El: Elbow|05263|0: Negative (Normal)|
|Vld: Valid Image|Fi: Finger||1: Positive (Abnormal)|
||Fo: Forearm|||
||Ha: Hand|||
||Hu: Humerus|||
||Sh: Shoulder|||
||Wr: Wrist|||
##### Then makes (Train.csv - Valid.csv) files for each bone type which will contain 3 columns without column headers or row index as the following sample:
|Row Index|Label for Normality|Image Path|Bone Type|
|--|--|--|--|
|0|1|ARNG\\Train\\Shoulder\\Trn-Sh-00001-1.png|Shoulder|

```python
import pandas as pd
import os, shutil
```

```python
def ArrFile(dirPath, dCSVName, IndexCSVName):
    dirPathLength= len(dirPath)
    st1= dCSVName[-25:-1]
    st2= st1[0:3]
    dType= ""                                                                 #Data Type Train/Valid
    # The no. of letters (tra/val) from right  direction is (25)
    # ImagePathes-train_labeled_studies.csv
    if st2=="tra":
        dType= "Trn"
        newPath= dirPath + "\\ARNG\\Train"
    else:
        dType= "Vld"
        newPath= dirPath + "\\ARNG\\Valid"
            
    dFllPath= dirPath + "\\" + dCSVName                                       #Data Full Path
    IndexFllPath= dirPath + "\\" + IndexCSVName                               #Index Full Path
    
    dData= pd.read_csv(dFllPath, sep= ',', names= ["Label", "Image Path", "Bone Type"])
    IndexNo= pd.read_csv(IndexFllPath, sep= ',', names= ["Index No"])
    
    # Elbow Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[0])
    p2= int(IndexNo.iloc[1]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Elbow Folder
    BonePath= newPath + "\\" + "Elbow"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Elbow Folder & Renaming to Trn-El-00001-1.png
    print('\n [INFO] copying images to elbow folder...')
    i=0
    # i from 0 to length of dfElbow -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-El-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Elbow List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Elbow-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file
    
    # Finger Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[2])
    p2= int(IndexNo.iloc[3]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Finger Folder
    BonePath= newPath + "\\" + "Finger"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Finger Folder & Renaming to Trn-Fi-00001-1.png
    print('\n [INFO] copying images to finger folder...')
    i=0
    # i from 0 to length of dfFinger -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Fi-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Finger List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Finger-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file
    
    # Forearm Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[4])
    p2= int(IndexNo.iloc[5]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Forearm Folder
    BonePath= newPath + "\\" + "Forearm"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Forearm Folder & Renaming to Trn-Fo-00001-1.png
    print('\n [INFO] copying images to forearm folder...')
    i=0
    # i from 0 to length of dfForearm -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Fo-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Forearm List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Forearm-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file
    
    # Hand Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[6])
    p2= int(IndexNo.iloc[7]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Hand Folder
    BonePath= newPath + "\\" + "Hand"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Hand Folder & Renaming to Trn-Ha-00001-1.png
    print('\n [INFO] copying images to hand folder...')
    i=0
    # i from 0 to length of dfHand -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Ha-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Hand List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Hand-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file
    
    # Humerus Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[8])
    p2= int(IndexNo.iloc[9]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Humerus Folder
    BonePath= newPath + "\\" + "Humerus"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Humerus Folder & Renaming to Trn-Hu-00001-1.png
    print('\n [INFO] copying images to humerus folder...')
    i=0
    # i from 0 to length of dfHumerus -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Hu-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Humerus List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Humerus-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file
    
    # Shoulder Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[10])
    p2= int(IndexNo.iloc[11]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Shoulder Folder
    BonePath= newPath + "\\" + "Shoulder"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Shoulder Folder & Renaming to Trn-Sh-00001-1.png
    print('\n [INFO] copying images to shoulder folder...')
    i=0
    # i from 0 to length of dfShoulder -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Sh-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Shoulder List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Shoulder-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file 
    
    # Wrist Data -------------------------------------------------------------
    p1= int(IndexNo.iloc[12])
    p2= int(IndexNo.iloc[13]) + 1
    Bone= dData.iloc[p1 : p2, :]
    # Making Wrist Folder
    BonePath= newPath + "\\" + "Wrist"
    if not os.path.exists(BonePath):
        os.makedirs(BonePath)
    
    # Copying Images to Wrist Folder & Renaming to Trn-Wr-00001-1.png
    print('\n [INFO] copying images to wrist folder...')
    i=0
    # i from 0 to length of dfWrist -1
    for i in range(len(Bone)):
    #for i in range(3):
        x= Bone.iloc[i,1]
        lb= str(Bone.iloc[i,0])
        oldFllPath= str(dirPath + "\\" + x)
        # This function replaces "/" by "\\"
        oldFllPath= oldFllPath.replace("/", "\\")
        j=format(i, '0>5')
        newFllPath= str(BonePath + "\\" + dType + "-Wr-" + j + "-" + lb + ".png")
        # This function replaces "/" by "\\"
        newFllPath= newFllPath.replace("/", "\\")
        shutil.copy(oldFllPath, newFllPath)
        letterNo= -1 *  (len(newFllPath) - dirPathLength - 1)
        Bone.iloc[i,1]= newFllPath[letterNo : ]
    
    print('\n [INFO] saving label and image as .csv file...')
    dfBone= pd.DataFrame(Bone)                                                  #Convert Wrist List to DataFrame
    # header: for column name, index: for raw name
    BoneFllPath= newPath + "\\Wrist-" + dCSVName
    dfBone.to_csv(BoneFllPath, sep=',', header= False, index=False)             #Save Label and Image as .csv file 
    
```
